```python
class dataset:
    col_names = ['Time','V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24',
                'V25','V26','V27','V28','Amount','Class']

    sample_data = pd.read_csv("data/Train1.csv", header = 0,names = col_names, nrows = train_size)
    sample_data.to_pickle('data/Train1.pkl')
    sample_data1 = pd.read_csv("data/Test1.csv", header = 0,names = col_names, nrows = test_size)
    sample_data1.to_pickle('data/Test1.pkl')
```

```python
def process(df):
    
    df = df.loc[:,('Time', 'V1','V2', 'V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24',
                'V25','V26','V27','V28','Amount','Class')].copy()
    return df
train = process(pd.read_pickle('data/Train1.pkl'))
test = process(pd.read_pickle('data/Test1.pkl'))
label = dataset.sample_data.Class.copy()

train
```

```python
category_variables = ["Class"]

for cv in category_variables:
    train[cv] = train[cv].astype("category")
    test[cv] = test[cv].astype("category", categories = train[cv].cat.categories)
```

```python
dummy_variables_labels = category_variables
    
class preprocessing:
    train_labels = pd.get_dummies(train, columns = dummy_variables_labels, prefix=dummy_variables_labels)
    test_labels = pd.get_dummies(test, columns = dummy_variables_labels, prefix=dummy_variables_labels)
```

```python
preprocessing.train_labels.to_pickle("data/preprocessed_train_data.pkl")
preprocessing.test_labels.to_pickle("data/preprocessed_test_data.pkl")

preprocessing.train_labels
```
